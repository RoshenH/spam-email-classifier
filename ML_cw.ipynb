```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
import seaborn as sns
```

```python
# loading dataset
Data = pd.read_csv('spambase.data', header=None)
```

```python
with open('./spambase.names') as spam:
 text = spam.read()
labels = re.findall(r'\n(\w*_?\W?):', text)

Data = pd.read_csv('./spambase.data', header=None, names=labels +['spam'])
```

```python
# droping duplicates
data=Data.drop_duplicates()
```

```python
sns.boxplot(data["capital_run_length_longest"])
```

```python
# applying Z-scoring to find and remove outliers
upper_limit = data["capital_run_length_total"].mean() + 3*data["capital_run_length_total"].std()
lower_limit = data["capital_run_length_total"].mean() - 3*data["capital_run_length_total"].std()

upper_limit1 = data["capital_run_length_longest"].mean() + 3*data["capital_run_length_longest"].std()
lower_limit1 = data["capital_run_length_longest"].mean() - 3*data["capital_run_length_longest"].std()

print("Upper limit_total",upper_limit)
print("lower limit total",lower_limit)

print("Upper limit  longest",upper_limit1)
print("lower limit longest ",lower_limit1)
```

```python
outliers = (data["capital_run_length_total"] > upper_limit) | (data["capital_run_length_total"] < lower_limit) | (data["capital_run_length_longest"] > upper_limit1) | (data["capital_run_length_longest"] < lower_limit1)
```

```python
new_data = data.loc[~outliers]
```

```python
sns.boxplot(new_data["capital_run_length_total"])
```

```python
sns.boxplot(new_data["capital_run_length_total"])
```

```python
# scaling the dataset
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
scaled_data=s.fit_transform(new_data)
```

```python
Spam_data=pd.DataFrame(data=scaled_data, columns= new_data.columns)
```

```python
Spam_data
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pc=pca.fit_transform(Spam_data)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('Variance')
plt.show
```

```python
X = new_data.iloc[:, :57].values
Y = new_data.iloc[:, 57].values
```

```python
# spliting the dataset 
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaling = StandardScaler()
X_train = scaling.fit_transform(X_train)
X_test = scaling.transform(X_test)
```

```python
# KNN without pca
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn.fit(X_train,Y_train)

Y_pred = knn.predict(X_test)

from sklearn.metrics import confusion_matrix,accuracy_score

print(accuracy_score(Y_test,Y_pred))
```

```python
# plotting confusion matrix
knn_matrix_nopca = confusion_matrix(Y_test, Y_pred)
plt.figure()
sns.heatmap(knn_matrix_nopca, annot=True,square=True)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix for KNN without PCA')
plt.show()
from sklearn.metrics import classification_report

print(classification_report(Y_test, Y_pred))
```

```python
# DT without PCA
from sklearn.tree import  DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy', random_state = 0)
dt.fit(X_train,Y_train)

Y_pred=dt.predict(X_test)
from sklearn.metrics import confusion_matrix,accuracy_score

print(accuracy_score(Y_test,Y_pred))
```

```python
dt_matrix_nopca = confusion_matrix(Y_test, Y_pred)

# Plot confusion matrix heatmap for Decision Tree
plt.figure()
sns.heatmap(dt_matrix_nopca, annot=True, square=True)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix for Decision Tree')
plt.show()
from sklearn.metrics import classification_report

print(classification_report(Y_test, Y_pred))
```
